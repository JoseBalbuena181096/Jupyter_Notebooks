```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
pesos=[100,150,150,200,250,300,325,400,415,500,600,1000]
```

```python
#Calcular la media
pesos_media=np.mean(pesos)
pesos_media
```

```python
#Como calcular la moda
stats.mode(pesos)
```

```python
#Calcular el rango
np.max(pesos)-np.min(pesos)
```

```python
#Calculo de la desviacion estandar
pesos_std=np.std(pesos)
pesos_std
```

Coeficiente de Simetr√≠a
El coeficiente de simetria (skewness) se define como el tercer momento dividido por la
desviacion estandar al cubo, tiene la siguiente formula:


```python
pow3=lambda x:x**3
```

```python
tercer_momento=lambda x:pow3(x-pesos_media)
```

```python
#la funcion sum suma todos los elementos
#map es una funcion que recibe como parametros una funcion 
# y una lista de valores , esta aplica la funcion para
#cada uno de los valores.
simetria_pesos=sum(map(tercer_momento,pesos))/(len(pesos)*pow3(pesos_std))
simetria_pesos
```

```python
#Una mejor forma es usar la funcion scipy.stats.skew
stats.skew(pesos)
```

Coeficiente de Curtosis 

El coeficiente de oblicuidad o curtosis se define como el cuarto momento dividido por la desviacion estandar a la cuarta, tiene la siguiente forma



generalmente se le resta 3 al cuarto momento portque la desviacion estandar tiene una curtosis de 3


```python
#Calcular la curtosis
pow4=lambda x:x**4
```

```python
cuarto_momento= lambda x:pow4(x-pesos_media)
```

```python
curtosis=(sum(map(cuarto_momento,pesos))/(len(pesos)*pow4(pesos_std)))-3
curtosis
```

```python
curtosis=(sum(map(cuarto_momento,pesos))/(len(pesos)*pow4(pesos_std)))-3
curtosis
```

```python
stats.kurtosis(pesos)
```
