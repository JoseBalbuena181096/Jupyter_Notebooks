```python
#Visualizacion de datos
#importamos el modulo pandas
#Cargar datos del csv boston_dataset.csv
import pandas as pd
df=pd.read_csv("/home/jose/data/boston_dataset.csv")
df.head()
```

```python
#Para poder mostrar la grafica es necesario importar el modulo matplotlib.pyplot a
import matplotlib.pyplot as plt
#obtener una grafica para visualizar la relacion entre el indice de
#de crimen y el valor_mediano de las casa
df.plot.scatter(x="INDICE_CRIMEN",y="VALOR_MEDIANO") 
```

```python
#Gracias a estar en un jupyter notebook podemos  simplemente imprimir
#la grafica sin plt.show() con la instruccion
%matplotlib notebook
#alpha es para la transparencia de los puntos para saber la consentracion de puntos
df.plot.scatter(x="INDICE_CRIMEN",y="VALOR_MEDIANO",alpha=0.2) 
```

```python
from pandas.plotting import scatter_matrix
sm=scatter_matrix(df, alpha=0.2,diagonal="kde")
```

```python
#Para poder mejorar la grafica y rotar los ejes 
sm=scatter_matrix(df, alpha=0.2,figsize=(14,14),diagonal="kde")
#Solution from
#http://stackoverflow.com/questions/32560932/how-to-customize-a-scatter-matrix-to-see-all-titles
#change  label rotation
[s.xaxis.label.set_rotation(45) for s in sm.reshape(-1)]
[s.yaxis.label.set_rotation(0) for s in sm.reshape(-1)]
#May need to offset label when rotating to prevent overlap of figure
[s.get_yaxis().set_label_coords(-1,0.5) for s in sm.reshape(-1)]
[s.get_xaxis().set_label_coords(-0.2,0) for s in sm.reshape(-1)]
#hide all ticks
[s.set_xticks(()) for s in sm.reshape(-1)]
[s.set_yticks(()) for s in sm.reshape(-1)];

```

```python
#creacion de grupos 
#Agrupamos por  ciudad y el valor mediano
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad.head()
```

```python
#Un diagrama de barras se utliza para comparar una variable entre distintos 
#tipos de grupos
#Grafico de barra para EL valor_medio_por_ciudad
valor_por_ciudad.head(10).plot.bar()
```

```python
#Hacerlo de manera horizontal
valor_por_ciudad.head(10).plot.barh()
```

```python
#graficos de lineas sirven para indicar tendencias temporales , evolucion
df.groupby("RATIO_PROFESORES").VALOR_MEDIANO.mean().plot.line()
```
