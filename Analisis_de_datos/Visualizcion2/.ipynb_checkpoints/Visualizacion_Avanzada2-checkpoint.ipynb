```python
#Importamos librerias para utilizar mapas
import cartopy.crs as ccrs
from cartopy.io import img_tiles
```

```python
%matplotlib notebook
df.VALOR_MEDIANO.plot.kde()
```

```python
#Utilizamos el primer quintil
primer_quintil=df.VALOR_MEDIANO.quantile(0.2)
primer_quintil
```

```python
#Utilizamos el cuarto
cuarto_quintil=df.VALOR_MEDIANO.quantile(0.8)
cuarto_quintil
```

```python
%matplotlib notebook
#Creamos una imagen de Google
imagery=img_tiles.GoogleTiles()
#API DE MATPLOTLIB
ax=plt.axes(projection=imagery.crs)
#Coordenadas del mapa
limites_mapa=(-71.38,-70.77,42.03,42.47)
ax.set_extent(limites_mapa)
ax.add_image(imagery,10)
#Para imprimir los datos menores del primer quintil 
df_primer_qt=df[df.VALOR_MEDIANO<primer_quintil]
#Para imprimir los datos mayores del cuarto quintil 
df_cuarto_qt=df[df.VALOR_MEDIANO>cuarto_quintil]
#Imprimimos mapas para cada quintil
#transform=ccrs.Geodetic() para transforlar LON Y LAT en pixeles
plt.plot(df_cuarto_qt.LON,df_primer_qt.LAT,transform=ccrs.Geodetic(),marker=".",
         markersize=10,color="red",linewidth=0,alpha=0.2)
plt.plot(df_primer_qt.LON,df_primer_qt.LAT,transform=ccrs.Geodetic(),marker=".",
         markersize=10,color="blue",linewidth=0,alpha=0.2)
```

```python
#importamos Bokeh 
import bokeh.plotting as bk
#Como estamos trabajando en un notebook hay que decirle a 
#bokeh que estamos en uno
bk.output_notebook()
```

```python

#bokeh es una libreria imperativa
#mostramos el indice crimen 
df.INDICE_CRIMEN
```

```python
#Son muchos datos por lo que usaremos un quintil
df["CRIMEN_QUINTIL"]=pd.qcut(df.INDICE_CRIMEN,5)
#Observamos las categorias de los quintiles para el 
#el indice de delincuencia
df.CRIMEN_QUINTIL.cat.categories
```

```python
#Importamos la paleta de colores
from bokeh.palettes import brewer
#asignamos colores por cada quintil
colors=brewer["Spectral"][len(df.CRIMEN_QUINTIL.unique())]
colors
```

```python
p=bk.figure(
    #Medida en pixeles
    plot_width=600,
    plot_height=600,
    title="Habitaciones vs Valor vivienda vs crimen"
)
#Para cada quintil poner los puntos con color diferente
for i,quintil in enumerate(df.CRIMEN_QUINTIL.cat.categories):
    #Filtramos para cada quintil
    df_q=df[df.CRIMEN_QUINTIL==quintil]
    #Damos caracteisticas de color unico a cada quintil
    p.scatter(df_q.N_HABITACIONES_MEDIO,df_q.VALOR_MEDIANO,color=colors[i]
             ,legend="({}-{})".format(quintil.left,quintil.right))
#Mostramos la grafica
bk.show(p)
```

```python
#Hacer histograma con bokeh
import numpy as np
hist,edges=np.histogram(df.VALOR_MEDIANO,bins=20)
```

```python
hist#El numero de la frecuencia de cada categoria
```

```python
#Los valores de cada categoria
edges
```

```python
#Creamos otra figura de bokeh
p1=bk.figure(title="Histograma valor viviendad",tools="save,hover",background_fill_color="#F1DDCC")
#Para crear bloques
#El valor maximo hist y el minimo 0 para la altura de bloques
#Left inicio primer elemento , fin penultimo elemento y right inicio segundo elemento y fin ultimo elemento
p1.quad(top=hist,bottom=0,left=edges[:-1],right=edges[1:],fill_color="#026560")
bk.show(p1)
```
