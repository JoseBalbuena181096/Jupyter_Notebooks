# Cargar datos a trav√©s de la funcion open

```python
#Abrir el archivo en modo lectura
data3=open(mainpath+"/"+"customer-churn-model/Customer Churn Model.txt","r")
```

```python
#
#Lista con el nombre de las columnas eliminando espacios en blanco y separando por comas
cols=data3.readline().strip().split(",")
#Definimos el tama√±o del las columas
n_cols=len(cols)
n_cols
cols
```

```python
#Creamos un diccionario para almacenar el dataframe
main_dict={}
#Recoremos cada elemento d elos comunas y a√±adimos al diccionario con una lista vacia como elemntos
for col in cols:
    main_dict[col]=[]
main_dict
```

```python
#iniciamos un contador
counter=0
#recorremos cada linea del documento
for line in data3:
    #Para cada linea eliminamos espacios en blanco y separamos por comas
    values=line.strip().split(",")
    #Para cada clave del diccionario  a√±adimos  asu clave un elemento correspondiente a su columna 
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter+=1
print("El data set tiene %d filas y %d columnas"%(counter,n_cols))
```

```python
df3=pd.DataFrame(main_dict)
df3.head()
```

# Lectura y escritura de ficheros.

```python
infile=mainpath+"/"+"customer-churn-model/Customer Churn Model.txt"
outfile=mainpath+"/"+"customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile,"r") as infile1:
    with open(outfile,"w")as outfile1:
        for line in infile1:
            fields=line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4=pd.read_csv(outfile,sep="\t")
df4.head()
```

# Leer datos desde una URL

```python
medals_url="http://winterolympicsmedals.com/medals.csv"
```

```python
#Cargar  csv desde una url 
medals_data=pd.read_csv(medals_url)
```

```python
medals_data.head()
```

```python
#Manejar  url
import urllib3
http=urllib3.PoolManager()
r=http.request("GET",medals_url)
print(r.status)
response=r.data
```

```python
response_str=str(response)
response_separate=response_str.split("\\n")
response_separate
```

```python
for i in range(len(response_separate)):
    response_separate[i]=response_separate[i].split(",")
```

```python
main_dic={}
for col in response_separate[0]:
    main_dic[col]=[]
main_dic   
```

```python
response_separate[0][len(response_separate[0])-1]
```

```python
for i in range(1,len(response_separate)):
    for j in range(len(response_separate[0])):
        main_dic[response_separate[0][j]].append(response_separate[i][j])
```

```python
for keys in main_dic.keys():
    print(len(main_dic[keys]))
```

```python
data5=pd.DataFrame(main_dic)
data5.head()
data5.rename(columns={
    'b"Year':"A√±os",
    'Medal':"Medallas"  
},inplace=True)
data5.head()
```
