```python
#Ejercicio de busqueda de alojamiento en Airbnb 
"""
Supongamos que somos un agente de Airbnb localizado en lisboa, y tenemos que atender peticiones 
de varios clientes.  Tenemos un archivo llamado airbnb.csv localizado en la carpeta data 
donde tenemos toda la imformacion de datos alojados en Airbnb Lisboa
"""
import pandas as pd
airbnb=pd.read_csv("/home/jose/data2/airbnb.csv")
```

```python
airbnb.head()
```

```python
 """
 En concreto el dataset tiene las siguientes variables:
 - room_id:el identificador de la propiedad
 - host_id:el identificador del dueño de la propiedad
 - room_type:tipo de propiedad
 - neighborhood: el barrio de Lisboa
 - reviews: El numero de opiniones
 - overall_satisfaction: Puntuacion media del apartamento
 - accommodates: El numero de personas que se pueden alojar en la propiedad
 - bedrooms: El numero de habitaciones
 - price: El precio (en euros) por noche
 """
```

```python
"""
CASO 1
Alicia va ir de viaje a Lisboa durante una semana con su marido 
y sus dos hijos. Esta buscando un apartamento con habitaciones 
separadas para sus padres e hijos. No les importan el precio
, simplemente quieren tener una experiencia agradable. Esto significa
que solo aceptan lugares con mas de 10 criticas (reviews) con una 
una puntuacion (overall_satisfaction) mayor de 4. Cuando seleccionamos
habitaciones para alicia tenemos que asegurarnos de peor a amejor puntuacion.
Para aquellas habilaciones que tienen la misma puntuacion mostrar la 
que tiene la mayor numero de criticas,mostrar tres alternativas
"""
```

```python
#Forma uno, con el metodo query
seleccion_alicia=airbnb.query("reviews>10 & overall_satisfaction>4 & bedrooms>=2")
seleccion_alicia.head()
```

```python
#Forma 2
seleccion_alicia=airbnb[(airbnb.reviews>10) &(airbnb.overall_satisfaction>4)&(airbnb.bedrooms>=2) ]
seleccion_alicia.head()
```

```python
#Podemos ordenar  la seleccion, primero por la puntuacion y luego por la critica de forma acendente
seleccion_alicia.sort_values(by=["overall_satisfaction","reviews"],ascending=False)
```

```python
#Mostramos los primeros 3 valores
seleccion_alicia.head(3)
```

```python
"""
CASO 2
Roberto es un casero que tiene una casa en Airbnb. De vez en cuando nos llama preguntando
sobre cuales son las criticas de su alojamiento. Hoy esta particularmente enfadado, ya que su
hermana Clara ha puesto su casa en Airbnb y Roberto quiere asegurarse que su casa tiene mas 
criticas que la de Clara. Tenemos que crear un dataFrame con las propiedades de ambos. Las
id de las propiedades (room_id) de Roberto y Clara son 97503 Y 90387 respectivamente. Finalmente 
guardamos este data frame como un esxel llamado robero.xls
"""
```

```python
#Primero del data frame  airbnb filtramos la columna room_id y los elemnto de las columnas con el metodo
#isin los elementos  97503 y 90387
roberto=airbnb[airbnb.room_id.isin([97503,90387])]
roberto
```

```python
roberto.to_excel("roberto.xls",index=False)
```

```python
roberto=pd.read_excel("roberto.xls")
roberto
```

```python
"""
CASO 3
Diana va a Lisboa a pasar 3 noches y quiere conocer a gente nueva. Tiene un presupuesto de
50 euros para su alojamiento. Debemos buscarle las 10 propiedades mas baratas, dandole
preferencia a aquellas que sean habitaciones compartidas (room_type==Shared room), y para 
aquellas viviendas compartidas debemos elegir aquellas con mayor puntuacion
"""
```

```python
#Forma 1
#Aplicamos el metodo query filtramos  por precio y tipo de habitacion
seleccion_diana=airbnb.query("price<=50/3 & room_type=='Shared room'")
seleccion_diana.head()
```

```python
#Forma 2
seleccion_diana=airbnb[(airbnb.price<=50/3)&(airbnb.room_type=="Shared room")]
seleccion_diana.head()
```

```python
#Ordenamos por precio de menor a mayor y por puntuacion de mayor a menor priorizando precio
seleccion_diana.sort_values(by=["price","overall_satisfaction"],ascending=[True,False]).head(10)
```
