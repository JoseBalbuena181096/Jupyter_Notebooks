```python
#Procesaremos los datos de los mas votados
df.sort_values(by="votes",ascending=False)
```

```python
#Procesamos por fracciones de votos
df.sort_values(by="fraction_votes",ascending=False)
```

```python
#Obtener la suma de votos por estado
#Primero agrupamos por estado
df.groupby("state")
#Adrupamos los votos por estado
df.groupby("state").votes
#Sumamos los votos por estado
df.groupby("state").votes.sum()
#Ordenar los votos de mayor a menor
df.groupby("state").votes.sum().sort_values(ascending=False)
#Ordenar los votos de mayor a menor por partido 
df.groupby(["state","party"]).votes.sum().sort_values(ascending=False)
```
