```python
#Pandas2
#TIPO DE DATOS 
#Data frame se puede declara con un diccionario
import pandas as pd
import numpy as np
datos={"Nombre":["Chalina","Marisol","Yolanda","Tina"],
      "Calificaciones":["100","90","100","80"],
       "Deportes":["Futbol","Natacion","Basquetbol","Beisbol"],
       "Materias":["Calculo","Metodos Numericos","Cocina","Quimica"]
      }
df=pd.DataFrame(datos)
df.head()
```

```python
print(df)
print(type(df))
```

```python
#Manipulacion de datos 
#Datos no validos
datos2={"Nombre":["Chalina","Marisol","Yolanda","N/A"],
      "Calificaciones":["100","90",np.nan,"80"],
       "Deportes":["Futbol","Natacion","Basquetbol","N/A"],
       "Materias":["Calculo","Metodos Numericos","N/A","Quimica"]
      }
df=pd.DataFrame(datos2)
df.head()
```

```python
#Descripcion de imformacion
df.info()
```

```python
#Estaditicas basicas de la informacion
df.describe()
```

```python
#Remplazar np.nan por cero
nuevo=pd.DataFrame(df)
nuevo=nuevo.replace(np.nan,"0")
nuevo
```

```python
#Si quiero quitar la linea
nuevo2=pd.DataFrame(df)
nuevo2.dropna(how='any',inplace=True)
nuevo2.head()
```

```python
#Eliminar registro buscando por columna
columna=df[df["Materias"]!="N/A"]
columna
```

```python
#Llenar con ceros en el data frame
nuevo3=pd.DataFrame(df)
nuevo3.fillna(0,inplace=True)
nuevo3
```

```python
nuevo.describe()
```

```python
#Convertir Numero enteros Calificaciones
nuevo3["Calificaciones"]=nuevo3.Calificaciones.astype(int)
nuevo3.describe()
```
